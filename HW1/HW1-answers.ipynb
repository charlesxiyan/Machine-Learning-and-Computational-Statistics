### Question 2.a

```python
### Since I have an issue in understanding the requirement of the question,
### two solutions are given for scenarios as follows.
### (1) reverse the order of words in a string with spaces to delimite thoese words; 
### (2) simply reverse characters in a string.

### (1)
def reverse_words(s):
    if not s:
        return s
    return ' '.join(s.split()[::-1])
```

```python
### Test
r1 = reverse_words('Monster Hunter!')
r2 = reverse_words(None)

print(f'Show outcome above:\n{r1}\n{r2}')
```

```python
### (2)
def reverse_words(s):
    if not s:
        return s
    return s[::-1]

### Test
r1 = reverse_words('Monster Hunter!')
r2 = reverse_words(None)

print(f'Show outcome above:\n{r1}\n{r2}')
```

### Question 2.d

```python
def count_chars(s):
    return {char: s.count(char) for char in set(s)}
```

### Question 2.f

```python
class BankAccount():
    def __init__(self):
        self.balance = 0 
        print(f'Account Initiated with Balance: ${self.balance}')

    def deposit(self, amount):
        self.balance += amount
        print(f'Current Balance: ${self.balance}')
    
    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
            print(f'Current Balance: ${self.balance}')
        else:
            print(f'Insufficient Balance: ${self.balance}')
```

```python
### Test
acc = BankAccount()
acc.withdraw(100)
acc.deposit(1000)
acc.withdraw(50)
```
