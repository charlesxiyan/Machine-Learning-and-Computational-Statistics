### Question 2.a

```python
def reverse_words(s):
    if not s:
        return s
    return s[::-1]
```

```python
### Test
r1 = reverse_words('Monster Hunter!')
r2 = reverse_words(None)

print(f'Show outcome above:\n {r1} and {r2}')
```

```python
def is_even_and_positive(x):
    return True if x > 0 and x % 2 == 0 else False
```

### Question 2.c

```python
def second_largest(lst):
    lst.sort()
    return lst[-2]
```

### Question 2.d

```python
def count_chars(s):
    return {char: s.count(char) for char in set(s)}
```

### Question 2.e

```python
def squares_of_evens(lst):
    return [num ** 2 for num in lst if num % 2 == 0]
```

### Question 2.f

```python
class BankAccount():
    def __init__(self):
        self.balance = 0 
        print(f'Account Initiated with Balance: ${self.balance}')

    def deposit(self, amount):
        self.balance += amount
        print(f'Current Balance: ${self.balance}')
    
    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
            print(f'Current Balance: ${self.balance}')
        else:
            print(f'Insufficient Balance: ${self.balance}')
```

```python
### Test
acc = BankAccount()
acc.withdraw(100)
acc.deposit(1000)
acc.withdraw(50)
```
